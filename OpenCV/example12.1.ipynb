```python
import glob
import cv2
from pprint import pprint
```

```python
def get_match_value(image, template):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    threshold, image = cv2.threshold(image, 0, 255, cv2.THRESH_OTSU)
    template = cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)
    threshold, template = cv2.threshold(template, 0, 255, cv2.THRESH_OTSU)

    h, w = image.shape
    template = cv2.resize(template, (w, h))

    result = cv2.matchTemplate(image, template, cv2.TM_CCOEFF)

    return result
```

```python
test_img = cv2.imread(r"G:\Program\Python\OpenCV\Chapter12\image\test2\1.bmp")

template_imgnames = []
for i in range(0, 10):
    template_imgnames.extend(glob.glob(f"G:\Program\Python\OpenCV\Chapter12\image\{i}\*"))

match_values = []

for template_imgname in template_imgnames:
    template_img = cv2.imread(template_imgname)
    match_values.append(get_match_value(test_img, template_img).item())

max_value = max(match_values)

print(max_value)
print(match_values)
pprint(template_imgnames)
print(match_values.index(max_value) // 10)
```
